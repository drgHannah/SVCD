# Load Image

```python
import matplotlib.pyplot as plt
from svcd.svcd import SVCDutils, SVCDSegmentation

ipath = "../img/kea.jpg"
image = SVCDutils.load_image(ipath)

# Plot RGB image
plt.imshow(image.transpose(1,2,0))
plt.show()
```

# Load Scribble

```python
scribble = "../img/kea_scribble.png"
scribble, colors = SVCDutils.load_segmentation_from_image(scribble)

# Plot scribbles
s,h,w = scribble.shape
fig, axs = plt.subplots(1,s,figsize=(20,5))
for i in range(s):
    axs[i].imshow(scribble[i])
    axs[i].set_title(f"Scribble {i} - Color {colors[i]}")
plt.show()
```

# Run SVCD Algorithm

```python
import numpy as np

svcd = SVCDSegmentation(max_iter=1000)
segmentation, energies = svcd(image, scribble)

# Plot result
segmentation_max = np.argmax(segmentation, axis = 0)
plt.imshow(segmentation_max)
plt.show()
plt.close()
```
